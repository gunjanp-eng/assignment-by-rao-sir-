```
# Usage : zcat mapped.bed.gz | python script.py > output.tsv
from collections import defaultdict
from sys import stdin,stdout

# A dict to store all datas  x, y and counts
all_data = defaultdict(lambda:(defaultdict(lambda:0)))

# Looping inside the input file
for line in stdin :
  # Storing the values inside a list
  row = line.strip().split("\t")
  # Finding centre for the undigested sequence
  m_center = (int(row[2]) + int(row[3]))/2
  # Finding centre for the digested sequence (some part of sequence is protected by the histones)
  f_center = (int(row[8]) + int(row[9]))/2
  # x_axis
  x = f_center - m_center
  # Total length after the digestion ,fragment length(y_axis)
  y = int(row[11])
  # Incrementing the x,y values
  all_data[x][y] += 1

# Looping inside the dict and getting each x ,y ,counts
for x in all_data :
  for y in all_data[x] :
    stdout.write(f"{x}\t{y}\t{all_data[x][y]}\n")
```
